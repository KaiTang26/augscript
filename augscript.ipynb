```python
import requests
import json
import csv
```

```python
class Corn:
  def __init__(self, judgementNo, varietyName):
    self.judgementNo = judgementNo
    self.varietyName = varietyName
```

```python
corns = []
# The API endpoint
getVarietyAuthorizeMainSearchUrl = "http://XXX.127.42.47:XXXX/SDSite/Home/GetVarietyAuthorizeMainSearch?JudgementNo=&VarietyNameLike=&JudgementYear=2017&ApplyCompanyLike=&CropID=%E7%8E%89%E7%B1%B3&IsTransgenosis=&IsExit=&JudgementRegionID=%E5%9B%BD%E5%AE%B6&introductionFilingType=&_search=false&nd=1727400607552&rows=200&page=1&sidx=JudgementYear+desc+%2CRegionID%2CJudgementNo&sord=desc"

# A GET request to the API
getVarietyResponse = requests.get(getVarietyAuthorizeMainSearchUrl)


# print(response.json()['rows'])
for row in getVarietyResponse.json()['rows']:
    newCorn = Corn(row['JudgementNo'], row['VarietyName'])
    corns.append(newCorn)
    
print(len(corns))
```

```python
outputData = []
for corn in corns:
    url = "http://XXX.127.42.47:XXXX/SDSite/Home/GetAnnouncementInfo"
    data = {
        "judgementNo": corn.judgementNo,
        "VarietyName": corn.varietyName,
    }
    try:
        response = requests.post(url, json=data)
        print(corn.judgementNo)
        json_object = json.loads(response.json())[0]
    
        cornDetails = {
            "审定编号": json_object['JudgementNo'],
            "作物名称": json_object['CropID'],
            "品种名称": json_object['VarietyName'],
            "申请者": json_object['ApplyCompany'],
            "育种者": json_object['BreedingCompany'],
            "品种来源": json_object['VarietySource'],
            "特征特性": json_object['VarietyCharacter'],
            "产量表现": json_object['OutputExpression'],
            "栽培技术要点": json_object['PlantRequirment'],
            "审定意见": json_object['JudgementSuggestion']
        }
    
        outputData.append(cornDetails)
    except:
        print("An exception occurred")
        continue
    
    
```

```python
print(len(outputData))
```

```python
with open('corn_records_all.csv', 'w', newline='') as csvfile:
    fieldnames = ["审定编号", "作物名称", "品种名称", "申请者", "育种者", "品种来源", "特征特性", "产量表现", "栽培技术要点", "审定意见"]
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(outputData)
```
